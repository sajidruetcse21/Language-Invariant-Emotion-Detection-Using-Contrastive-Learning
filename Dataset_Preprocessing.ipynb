```python
# Install required library (usually already installed in Colab)
!pip install pandas scikit-learn openpyxl

import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# -----------------------------
# 1. Load Excel Dataset
# -----------------------------
file_path = "/content/Bengali_Banglish_80K_Dataset.xlsx"   # change if needed

df = pd.read_excel(file_path)

print("Dataset shape:", df.shape)
print(df.head())
```

```python
# -----------------------------
# 2. Check label distribution
# -----------------------------
label_column = "Label"   # change if your column name is different

print("\nClass distribution:")
print(df[label_column].value_counts())
```

```python
# -----------------------------
# 3. Stratified Split
# -----------------------------
# First split: Train (70%) + Temp (30%)

train_df, temp_df = train_test_split(
    df,
    test_size=0.30,
    stratify=df[label_column],
    random_state=42
)

# Second split: Validation (10%) + Test (20%)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.67,   # because 20/(10+20) ≈ 0.67
    stratify=temp_df[label_column],
    random_state=42
)
```

```python
# -----------------------------
# 4. Verify sizes
# -----------------------------
print("\nSplit sizes:")
print("Train:", len(train_df))
print("Validation:", len(val_df))
print("Test:", len(test_df))
```

```python
# -----------------------------
# 5. Verify class distribution
# -----------------------------
print("\nTrain distribution:")
print(train_df[label_column].value_counts())

print("\nValidation distribution:")
print(val_df[label_column].value_counts())

print("\nTest distribution:")
print(test_df[label_column].value_counts())
```

```python
# -----------------------------
# 6. Save CSV files
# -----------------------------
train_df.to_csv("/content/train.csv", index=False, encoding="utf-8-sig")
val_df.to_csv("/content/validation.csv", index=False, encoding="utf-8-sig")
test_df.to_csv("/content/test.csv", index=False, encoding="utf-8-sig")

print("\nFiles saved:")
print("train.csv")
print("validation.csv")
print("test.csv")
```
